```python
import pandas as pd
import numpy as np
```

```python
dodge = pd.read_csv('edmundsconsumer-car-ratings-and-reviews/Scraped_Car_Review_dodge.csv')
dodge.head()
```

```python
dodge.shape
```

```python
dodge.columns
```

```python
dodge.rename(columns={'Unnamed: 0': 'unnamed', 'Review_Date': 'review_data', 'Author_Name': 'author_name',
                    'Vehicle_Title': 'vehicle_title', 'Review_Title': 'review_title', 'Review': 'review',
                    'Rating': 'rating'}, inplace=True)
```

```python
dodge.head()
```

```python
dodge = dodge.dropna()
```

```python
dodge['vehicle_year'] = dodge['vehicle_title'].str.extract('(\d\d\d\d)', expand=True)
dodge.head()
```

```python
dodge['review_date'] = dodge['review_data'].str.extract('(\d\d/\d\d/\d\d)', expand=True)
dodge.head()
```

```python
dodge = dodge.drop(['review_data'], axis=1)
dodge.head()
```

```python
dodge = dodge.drop(['unnamed'], axis=1)
dodge.head()
```
